# Introducción
Pac-Man (en esta versión simplificada) se trata de un juego en el que un personaje (**Mr. Pacman**)se desplaza por un tablero, en el que tenemos casillas libres y paredes o muros. 
En este tablero, en las casillas libres hay comida. Cuando pacman se coloca en una casilla con alimento, el alimento desaparece y se incrementan los puntos.

<img style="float:left" width="40%" src="pics/comecocos.png">



## Partes del juego

- __Tablero__: Define las partes del suelo libres o que están ocupadas por un muro. El tablero es inmutable, no cambia nunca.
- __MrPacman__: Define la posición del personaje en términos de coordenadas y,x (fila, columna), su posición sí que cambia.
- __Comida__: Algunas casillas tienen comida. Cuando Mr Pacman pisa una de esas casillas, desaparece la comida y aumentan los puntos. La comida, por lo tanto, solo puede ir desapareciendo.
- __Puntos__: Un número que se va incrementando a medida que se come la comida del tablero.

## ¿ Que tipo de dato sería cada parte del juego atendiendo a los datos anteriores?

- **Tablero**. El tablero no va a cambiar nunca, podría ser una lista de listas. Cada lista interna representando una fila. Y cada valor de la fila representaría con 0 que es una casilla libre, o con un 1, que hay un muro.
- **MrPacman** va a cambiar. Queremos hacer operaciones de movimiento que lo desplacen hacía arriba, abajo, derecha e izquierda. Por tanto, lo mejor es que sea una lista de coordenadas.
- **Comida**: Se quiere saber si una determinada coordenada está entre las que tiene comida y se quieren eliminar elementos cuando la comida se consuma. Comida será una lista de listas y cada lista interna será una coordenada, que indica que en dicha coordenada hay comida. Usando una lista de coordenadas se puede hacer lo primero muy facilmente con el operador **in** y lo segundo con **remove**.
- **Puntos**. Los puntos son una variable numérica.


Vamos a crear una clase JuegoPacman que nos permita tener y trabajar con todos estos elementos.

Para ello, vamos a ir paso a paso generando los diferentes elementos y por último, tendrás que implementar la clase poniendo de forma ordenada todos los elementos.

direccion = None
terminado = False
def on_press(tecla):
    nonlocal direccion, terminado
    try:
        if tecla == keyboard.Key.esc:
            terminado = True
            return False  # Detener el listener
        elif tecla == keyboard.Key.up:
            direccion = arriba
        elif tecla == keyboard.Key.down:
            direccion = abajo
        elif tecla == keyboard.Key.left:
            direccion = izquierda
        elif tecla == keyboard.Key.right:
            direccion = derecha
    except:
        pass

listener = keyboard.Listener(on_press=on_press)
listener.start()

### Emplea este código para poder jugar de forma visual

El widget HTML es capaz de mostrar HTML.
Una de las cosas que puede hacer es mostrar una tabla HTML con una imagen diferente por cada tipo de elemento.

Si todo está implementado correctamente, la ejecución de la siguiente celda, debería mostrarte un tablero con todos los elementos.
Puede que encuentres algunas pistas en este código que te permitan ajustar la creación de la clase `JuegoPacman`.

````python
from proyecto_pacman.utils import arriba, abajo, izquierda, derecha, tableros
from IPython.display import display
from ipywidgets import HTML ,Button, HBox, VBox
from proyecto_pacman.clases import JuegoPacman 

def on_button_clicked(b):
    global juegoPacman
    desc=b.description
    mov = [0,0]
    if desc=="^":
        mov=arriba
    elif desc=="v":
        mov=abajo
    elif desc==">":
        mov=derecha
    elif desc=="<":
        mov=izquierda
    
    juegoPacman.actualizaJuego(mov)
    visor.value = pinta_juegoHTML(juegoPacman)
    if juegoPacman.comida_agotada():
        visor.value =final_juegoHTML(juegoPacman.get_puntos())
        
    
def pinta_juegoHTML(juegoPacman):

    cadenaHTML = '<table cellspacing="0" cellpadding="0" style="border-collapse: collapse;"><tr>'

    # Plantillas de imágenes sin bordes ni separación
    img_tpl = '<td style="padding:0; margin:0;"><img src="{}" height="40" width="40" style="display:block;"></td>'
    
    # Rutas a las imágenes
    img_map = {
        "#": "./spritesPacman/block.jpg",
        "@": "./spritesPacman/pacman.jpg",
        ".": "./spritesPacman/comida.jpg",
        " ": "./spritesPacman/hueco.jpg"
    }
    juego_str=str(juegoPacman)
    lineas = juego_str.splitlines()
    for i, linea in enumerate(lineas):
        for j, caracter in enumerate(linea):
            src = img_map.get(caracter, img_map[" "])
            cadenaHTML += img_tpl.format(src)
        cadenaHTML += "</tr><tr>"  # fin de fila y comienzo de la siguiente
    cadenaHTML += "</tr></table>"
    cadenaHTML += "<div style='margin-top:10px;'>Puntos: {}</div>".format(juegoPacman.get_puntos())
    
    return cadenaHTML

def final_juegoHTML(puntos):
    cadenaHTML = "<div style='margin-top:10px;'>El juego ha terminado. Has ganado con {} puntos</div>".format(puntos)
    return cadenaHTML



########################
## JUEGO
########################
# En cada partida se rellena el tablero de comida (cambiando "easy", "medium", "hard", tendrás acceso a los diferentes tableros)
juegoPacman=JuegoPacman(tableros["hard"],[1,1])       

# visor HTML donde se representará el juego
visor=HTML()
visor.value = pinta_juegoHTML(juegoPacman)

# Botones para las direcciones
up = Button(description="^")
down = Button(description="v")
right = Button(description=">")
left = Button(description="<")

empty=Button(description=" ")
empty.margin=2

# se añaden eventos a los botones
up.on_click(on_button_clicked)
down.on_click(on_button_clicked)
right.on_click(on_button_clicked)
left.on_click(on_button_clicked)

# Estética: Se crea una caja vertical con dos cajas horizontales dentro
# Tenemos 2 x 3 casillas para colocar los botones y los colocamos en forma de teclado
control=VBox([HBox([empty,up,empty]),HBox([left,down,right])])

# El juego es el visor y los controles
ui=VBox(children=[visor, control])

display(ui)
````
